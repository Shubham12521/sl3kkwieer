```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
print(df)
```

```python
sns.displot(df['fare'])
```

```python
sns.jointplot(x='age', y='fare', data=df)
```

```python
sns.catplot(x="sex", hue="survived", kind="count", data=df)
```

```python
group = df.groupby(['pclass', 'sex'])
pclass_sex = group.size().unstack()
sns.heatmap(pclass_sex, annot=True, fmt="d")
```

```python
sns.barplot(x="sex", y="age", hue="survived", data=df)
```

```python
sns.catplot(x='embarked', hue='survived', kind='count', col='pclass', data=df)
```
